# Question 1

```python
transactions = [1200, 3000, 250, 7800, 540, 3000, 9100, 250, 1500, 7800]
duplicate_transactions = [] 
for i in range(len(transactions)):
    count = 0
    for j in range(len(transactions)):
        if transactions[i] == transactions[j]:
            count = count + 1
    if count > 1:
        already_added = False
        for d in duplicate_transactions:
            if d == transactions[i]:
                already_added = True
        if not already_added:
            duplicate_transactions.append(transactions[i])
#Duplicate transactions                
print(duplicate_transactions)

#Fraud monitoring for transactions more than 5000
for t in transactions:
    if t > 5000:
        print("The amount of", t,  "may require Fraud monitoring")

# Total sum of transactions

print("Total spending of the day is: ", sum(transactions))

# Unique transactions list

unique_transactions = []
for t in transactions:
    is_present = False
    for u in unique_transactions:
        if u == t:
            is_present = True
    if not is_present:
        unique_transactions.append(t)
print("The unique transaction list is", sorted(unique_transactions))


# Unique transactions list descending order

desc_unique_transactions= unique_transactions

desc_unique_transactions.sort(reverse = True)

print("The descending order of the unique transactions:", desc_unique_transactions)
```

# Question 2

```python
ages = [25, 42, -1, 34, None, 52, 17, -5, 29, None, 46]
valid_ages    = []
invalid_count = 0
none_count    = 0
negative_count = 0
 
for age in ages:
    if age is None:
        none_count    = none_count + 1
        invalid_count = invalid_count + 1
    elif age < 0:
        negative_count = negative_count + 1
        invalid_count  = invalid_count + 1
    else:
        valid_ages.append(age)

#Remove invalid ages and store valid ages
print("Valid Ages  :", valid_ages)
#print("Valid Count :", len(valid_ages))

# Calculate average age on valid ages
valid_count = len(valid_ages)
sum_valid_ages = sum(valid_ages)

avg_valid_age = sum_valid_ages / valid_count

print("The average valid age:", avg_valid_age)


# Displays the list of customers whose age is greater than 30
age_grt_30 = []
for age in valid_ages:
    if age > 30:
        age_grt_30.append(age)

age_grt_30.sort()
print("The list of age greater than 30:", age_grt_30)

# Number of invalid ages

print("The number of invalid ages:", invalid_count)
```

# Question 3

```python
products = ["Laptop", "Phone", "Tablet", "Monitor", "Headphones"]
amazon_prices = [75000, 45000, 30000, 20000, 5000]
flipkart_prices = [72000, 47000, 31000, 19500, 4800]

table = []
for i in range(len(products)):
    row = [products[i], amazon_prices[i], flipkart_prices[i]]
    table.append(row)

for row in table:
    product_name = row[0]
    amazon_price = row[1]
    flipkart_price = row[2]

    if amazon_price < flipkart_price:
        cheaper_platform = "Amazon"
        difference = flipkart_price - amazon_price
    elif flipkart_price < amazon_price:
        cheaper_platform = "Flipkart"
        difference = amazon_price - flipkart_price
    else:
        cheaper_platform = "Both same price"
        difference = 0

    print("Product:", product_name)
    print("Cheaper platform:", cheaper_platform)
    print("Price difference:", difference)
    print("--------------------------")
```

# Question 4

```python
login_hours = [1, 3, 5, 7, 2, 6, 8, 4, 9, 10] 
even_logins=[]
odd_logins=[]
for hours in login_hours:
    if hours%2 == 0:
        even_logins.append(hours)
    else:
        odd_logins.append(hours)
print("The even login hours are:", even_logins)
print("The odd login hours are:", odd_logins)

#Peak activity time
peak_login=[]
for hours in login_hours:
    if hours > 5:
        peak_login.append(hours)
print("The peak login hours recorded are:", peak_login)

#Sort Login hours
sort_login_hours = login_hours

sort_login_hours.sort()

print("The sorted login hours is:", sort_login_hours)

#Count number of peak logins 

count_peak= len(peak_login)

print("The number of peak logins are:", count_peak)
```

# Question 5

```python
orders = [
    ["ORD001", "Laptop",   75000, "Completed"],
    ["ORD002", "Mouse",      500, "Completed"],
    ["ORD003", "Laptop",   75000, "Returned"],
    ["ORD004", "Keyboard",  1500, "Completed"],
    ["ORD005", "Laptop",   75000, "Completed"],
    ["ORD006", "Monitor",  20000, "Fraud"],
    ["ORD007", "Mouse",      500, "Completed"],
    ["ORD008", "Laptop",   75000, "Completed"],
]

HEADER = ["Order_ID", "Product_Name", "Price", "Status"]

for col in HEADER:
    print(f"{col:<15}", end="")
print()

for order in orders:
    for field in order:
        print(f"{str(field):<15}", end="")
    print()  

print()


# For Fraud orders identification 

fraud_found = False
for order in orders:
    if order[3] == "Fraud":
        fraud_found = True
        print(f"{order[0]}, Status is fraud" )
if not fraud_found:
    print ("No Fraud found")

all_products = []
for order in orders:
    all_products.append(order[1])

unique_products = []

suspicious_products = []

for product in all_products:
    if product not in unique_products:
        unique_products.append(product)

for product in unique_products:
    count = 0
    for order in orders:
        if order[1] == product:
            count += 1
    if count > 3:
        print(f"{product} appears {count} times — flagged as suspicious!")
        suspicious_products.append(product)

if not suspicious_products:
    print("  No product exceeds more than 3 time.")

print()

# Total revenue generated

total_revenue = 0
for order in orders:
    if order[3] == "Completed":
        total_revenue = total_revenue + order[2]

print(f"Total Revenue for the day: {total_revenue:,}")
print()


# Total count of completed returned and fraud orders
completed_count = 0
returned_count = 0
fraud_count = 0

for order in orders:
    if order[3] == "Completed":
        completed_count = completed_count + 1
    elif order[3] == "Returned":
        returned_count = returned_count + 1
    elif order[3] == "Fraud":
        fraud_count = fraud_count + 1

print(f"Completed : {completed_count}")
print(f"Returned  : {returned_count}")
print(f"Fraud     : {fraud_count}")


#Total number of unique products
print(f"{len(unique_products)} Unique products : {unique_products}")
print()

#Total revenue generated product based with completed status

for product in unique_products:
    product_orders = 0
    product_revenue = 0
    for order in orders:
        if order[1] == product:
            product_orders = product_orders + 1
            if order[3] == "Completed":
                product_revenue =  product_revenue + order[2]
    print(f"{product} {product_orders} {product_revenue:}")

print()

# Most sold product and count of orders
most_sold_product = ""
max_count = 0

for product in unique_products:
    count = 0
    for order in orders:
        if order[1] == product:
            count = count + 1
    if count > max_count:
        max_count = count
        most_sold_product = product

print(f"Most Sold Product : {most_sold_product}")
print(f"Number of Orders  : {max_count}")
print()

#Remove fraud order and display clean dataset
cleaned_orders = []
for order in orders:
    if order[3] != "Fraud":
        cleaned_orders.append(order)

print (cleaned_orders)
print()
```
