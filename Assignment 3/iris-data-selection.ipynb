# Data Selection for Model Input — Iris Dataset

Models are trained on **selective features**, not full raw data.  
This notebook covers NumPy slicing, boolean masking, and their role in ML pipelines.

**Tasks:**
1. Extract first 100 samples + last 2 features
2. Boolean masking — select samples where petal length > dataset mean
3. Count selected samples
4. Relate this to feature-based filtering in ML

```python
import numpy as np
from sklearn.datasets import load_iris

iris = load_iris()
X = iris.data          # shape (150, 4)
y = iris.target        # shape (150,)
feature_names = iris.feature_names

print("Full dataset shape :", X.shape)
print("Feature names      :", feature_names)
print("Classes            :", iris.target_names)
```

```python
# Select first 100 rows AND last 2 columns simultaneously
X_selected = X[:100, -2:]

print("Selected features  :", feature_names[-2:])   # petal length, petal width
print("Shape after slicing:", X_selected.shape)      # (100, 2)
print()
print("First 100 rows:")
print(X_selected[:100])
print()
print("Last 2 rows:")
print(X_selected[-2:])
```

```python
# Petal length is column index 2 in the full dataset
petal_length = X[:, 2]

# Compute dataset-wide mean of petal length
mean_petal_length = np.mean(petal_length)
print(f"Mean petal length : {mean_petal_length:.4f} cm")

# --- Build the boolean mask ---
mask = petal_length > mean_petal_length

print(f"\nMask shape        : {mask.shape}")
print(f"Mask dtype        : {mask.dtype}")
print(f"First 100 mask vals: {mask[:100]}")
print(f"  (False = below mean, True = above mean)")

print(f"Count of selected sample: {np.sum(mask)}")
```
