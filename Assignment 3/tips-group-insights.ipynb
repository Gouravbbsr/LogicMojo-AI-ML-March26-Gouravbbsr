# Q8. Group-Based Insights (Customer Segmentation Logic)
**Context:** ML models often depend on aggregated patterns.
**Dataset:** Tips

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')
print('Shape:', tips.shape)
tips.head()
```

## Task 1a: Average Tip by Day

```python
avg_tip_by_day = tips.groupby('day')['tip'].mean().round(2)
print('Average Tip by Day:')
print(avg_tip_by_day)
```

## Task 1b: Total Bill by Gender

```python
total_bill_by_gender = tips.groupby('sex')['total_bill'].agg(['sum', 'mean', 'count']).round(2)
total_bill_by_gender.columns = ['Total Bill ($)', 'Avg Bill ($)', 'Num Transactions']
print('Total Bill Summary by Gender:')
print(total_bill_by_gender)
```

## Task 2: Create New Feature — tip_percentage

```python
tips['tip_percentage'] = (tips['tip'] / tips['total_bill'] * 100).round(2)

print('tip_percentage stats:')
print(tips['tip_percentage'].describe().round(2))
print()
print('Sample rows with new feature:')
tips[['total_bill', 'tip', 'tip_percentage']].head(8)
```

## Task 3: Which Day Has the Highest Tipping Behavior?

```python
day_tipping = tips.groupby('day').agg(
    avg_tip_pct=('tip_percentage', 'mean'),
    avg_tip_amt=('tip', 'mean'),
    num_customers=('tip', 'count')
).round(2)

best_day = day_tipping['avg_tip_pct'].idxmax()
print(f'Day-wise Tipping Summary:')
print(day_tipping)
print(f'\nHighest Tipping Day: {best_day} ({day_tipping.loc[best_day, "avg_tip_pct"]}% avg tip)')
```
