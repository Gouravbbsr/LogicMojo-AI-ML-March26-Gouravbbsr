# Q9. Advanced Filtering + Subsetting
**Context:** Training data is often filtered based on business rules.
**Dataset:** Tips

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')
print('Original dataset shape:', tips.shape)
tips.head()
```

## Task 1: Filter — total_bill > 20 AND tip < 3

```python
mask = (tips['total_bill'] > 20) & (tips['tip'] < 3)
filtered_df = tips[mask].copy()

filtered_df
```

## Task 2: Select Only Relevant Columns

```python
relevant_cols = ['total_bill', 'tip', 'sex', 'day', 'time', 'size']
subset = filtered_df[relevant_cols].copy()


print('Subset shape:', subset.shape)
subset
```

## Task 3: Analyze — Is This Segment Under-Tipping?

```python
# Compute tip % for full dataset baseline
subset['tip_percentage'] = (subset['tip'] / subset['total_bill'] * 100).round(2)

tips['tip_percentage'] = (tips['tip'] / tips['total_bill'] * 100).round(2)

overall_avg_tip_amt  = tips['tip'].mean()
segment_avg_tip_amt  = subset['tip'].mean()

print('=== Under-Tipping Analysis ===')
print(f'Overall avg tip ($) : ${overall_avg_tip_amt:.2f}')
print(f'Segment avg tip ($) : ${segment_avg_tip_amt:.2f}')

if segment_avg_tip_amt < overall_avg_tip_amt:
    print(f'\nThis segment IS under-tipping by {overall_avg_tip_amt - segment_avg_tip_amt:.2f}')
else:
    print('\nThis segment is tipping normally or above average.')
```
