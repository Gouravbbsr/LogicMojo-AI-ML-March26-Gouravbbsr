# Q10. Data Transformation for Model Input (NumPy)
**Context:** ML models require structured numeric arrays.

```python
import numpy as np

# Task 1: Create two arrays of shape (3, 2)
A = np.array([[1, 2],
              [3, 4],
              [5, 6]])

B = np.array([[7,  8],
              [9, 10],
              [11, 12]])

print('Array A (3x2):')
print(A)
print('\nArray B (3x2):')
print(B)
```

## Task 2a: Vertical Stacking — `np.vstack`
Stacks arrays row-by-row. Both arrays must have the **same number of columns**.

```python
v_stacked = np.vstack([A, B])
print('Vertical Stack (vstack) shape:', v_stacked.shape)  # (6, 2)
print(v_stacked)
```

## Task 2b: Horizontal Stacking — `np.hstack`
Stacks arrays column-by-column. Both arrays must have the **same number of rows**.

```python
h_stacked = np.hstack([A, B])
print('Horizontal Stack (hstack) shape:', h_stacked.shape)  # (3, 4)
print(h_stacked)
```

## Task 3: Reshape vstack result into (2, 6)

```python
reshaped = v_stacked.reshape(2, 6)
print('Original vstack shape :', v_stacked.shape)  # (6, 2) → total elements = 12
print('Reshaped to (2, 6)     :', reshaped.shape)   # (2, 6) → total elements = 12
print()
print('Reshaped array:')
print(reshaped)
print()
# verify no data was lost
print('Total elements before:', v_stacked.size)
print('Total elements after :', reshaped.size)
```
