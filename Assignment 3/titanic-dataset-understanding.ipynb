The **first step** in any ML pipeline is dataset inspection.  
You cannot build a good model without understanding what's in your data.

**Tasks covered:**
1. `head()`, `tail()`, `info()`, `describe()`
2. Missing values per column
3. Numerical vs categorical features
4. Why feature types matter before modeling

```python
import pandas as pd
import numpy as np
import seaborn as sns

# Load Titanic dataset from seaborn
df = sns.load_dataset('titanic')

print("Dataset loaded successfully!")
print(f"Shape: {df.shape}  →  {df.shape[0]} rows, {df.shape[1]} columns")
print(f"Columns: {list(df.columns)}")
```

```python
total_rows = len(df)

missing = pd.DataFrame({
    'Missing Count'  : df.isnull().sum(),
    'Missing %'      : (df.isnull().sum() / total_rows * 100).round(2),
    'Dtype'          : df.dtypes
})

# Sort by missing count descending, show only columns with any missing
missing_only = missing[missing['Missing Count'] > 0]

print(f"Total rows: {total_rows}")
print(f"Columns with missing values: {len(missing_only)}")
```

## Task 2B — Numerical vs Categorical Features

```python
# Separate by dtype
numerical_cols   = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_cols = df.select_dtypes(include=['object', 'category', 'bool']).columns.tolist()

print(f"NUMERICAL features ({len(numerical_cols)}):")
for col in numerical_cols:
    print(f"  {col:<15} ")

print()
print(f"CATEGORICAL features ({len(categorical_cols)}):")
for col in categorical_cols:
    unique_vals = df[col].dropna().unique()
    preview = str(list(unique_vals[:5]))
    print(f"  {col:<15}")
```
