# Q15. Business + ML Insight Problem (Tips Dataset)
**Context:** Convert data analysis into decision-making

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

tips = sns.load_dataset('tips')
print('Shape:', tips.shape)
tips.head()
```

## Task 1: Create tip_percentage

```python
tips['tip_percentage'] = (tips['tip'] / tips['total_bill'] * 100).round(2)

print('tip_percentage sample stats:')
print(tips['tip_percentage'].describe().round(2))

tips[['total_bill', 'tip', 'tip_percentage']].head(6)
```

## Task 2: Group by Day and Time

```python
grouped = tips.groupby(['day', 'time']).agg(
    total_revenue   =('total_bill', 'sum'),
    total_tips      =('tip', 'sum'),
    avg_bill        =('total_bill', 'mean'),
    avg_tip_pct     =('tip_percentage', 'mean'),
    num_customers   =('tip', 'count')
).round(2)

grouped['revenue_per_customer'] = (grouped['total_revenue'] / grouped['num_customers']).round(2)
grouped = grouped.sort_values('total_revenue', ascending=False)

print('Revenue Summary by Day + Time:')
grouped
```

## Task 3: Identify Highest Revenue Segment

```python
best_segment     = grouped['total_revenue'].idxmax()
best_tip_segment = grouped['avg_tip_pct'].idxmax()

print(f'Highest total revenue: {best_segment}  →  ${grouped.loc[best_segment, "total_revenue"]:.2f}')
```
