# Q14. Error Analysis Simulation (Model Evaluation Thinking)
**Context:** Understanding prediction errors is key in ML.

```python
import numpy as np
import matplotlib.pyplot as plt

# Task 1: Generate 1000 random error values
np.random.seed(42)
errors = np.random.normal(loc=0, scale=1.5, size=1000)

print('Error array preview (first 10):')
print(np.round(errors[:10], 3))
print(f'\nTotal error samples: {len(errors)}')
```

## Task 2: Compute Mean and Standard Deviation

```python
mean_err = np.mean(errors)
std_err  = np.std(errors)

print('=== Error Statistics ===')
print(f'Mean (bias)         : {mean_err:.4f}')   
print(f'Std Dev (spread)    : {std_err:.4f}')

```

## Task 3: Identify Outliers — Using Mean ± 2*std

```python
lower_bound = mean_err - 2 * std_err
upper_bound = mean_err + 2 * std_err



print(f'Outlier bounds  : [{lower_bound:.3f}, {upper_bound:.3f}]')
```
