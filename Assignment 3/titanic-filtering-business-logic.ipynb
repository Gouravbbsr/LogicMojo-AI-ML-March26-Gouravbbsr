# Filtering Data for Business Logic — Titanic

Real ML problems require filtering **relevant populations** before analysis.  
Applying domain knowledge to filter data is the bridge between raw data and meaningful features.

**Tasks:**
1. Filter — Female passengers in 1st class
2. Compute — Survival rate for this group
3. Compare — With overall survival rate
4. Interpret — ML feature engineering insights

```python
import pandas as pd
import numpy as np
import seaborn as sns

df = sns.load_dataset('titanic')

print(f"Dataset shape : {df.shape}")
print()

# --- Build the boolean mask ---
mask_female   = df['sex']    == 'female'
mask_1st_cls  = df['pclass'] == 1

# Combined filter
mask_target   = mask_female & mask_1st_cls
df_target     = df[mask_target].copy()

print(f"Total passengers              : {len(df)}")
print(f"Female AND 1st class          : {len(df_target)}")
print()
```

```python
# --- Survival rate for female 1st-class passengers ---
target_survival_rate = df_target['survived'].mean()
target_survived      = df_target['survived'].sum()
target_died          = len(df_target) - target_survived

print("GROUP: Female passengers in 1st class")
print(f"  Total in group     : {len(df_target)}")
print(f"  Survived           : {int(target_survived)}")
print(f"  Did not survive    : {int(target_died)}")
print(f"  Survival rate      :{target_survival_rate*100:.1f}")
```

```python
# Overall survival rate
overall_rate = df['survived'].mean()


print(f"Overall survival rate (all passengers): {overall_rate*100:.1f}%")
print()
```
