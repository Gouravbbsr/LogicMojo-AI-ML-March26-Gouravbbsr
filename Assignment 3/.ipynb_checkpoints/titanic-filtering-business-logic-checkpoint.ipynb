# Filtering Data for Business Logic — Titanic

Real ML problems require filtering **relevant populations** before analysis.  
Applying domain knowledge to filter data is the bridge between raw data and meaningful features.

**Tasks:**
1. Filter — Female passengers in 1st class
2. Compute — Survival rate for this group
3. Compare — With overall survival rate
4. Interpret — ML feature engineering insights

````python
import pandas as pd
import numpy as np
import seaborn as sns

df = sns.load_dataset('titanic')

print(f"Dataset shape : {df.shape}")
print(f"Columns       : {list(df.columns)}")
print()
df[['survived', 'pclass', 'sex', 'age', 'fare', 'embarked']].head()
````

---
## Task 1 — Filter: Female Passengers in 1st Class

We combine **two boolean conditions** using `&` (AND).  
Each condition produces a boolean Series; `&` keeps only rows where **both** are True.

```
condition_1:  sex == 'female'   →  [True, False, True, ...]
condition_2:  pclass == 1       →  [True, True,  False, ...]
combined  :   cond_1 & cond_2   →  [True, False, False, ...]
```

> **Rule**: Always wrap each condition in parentheses when combining with `&` or `|`.

````python
# --- Build the boolean mask ---
mask_female   = df['sex']    == 'female'
mask_1st_cls  = df['pclass'] == 1

# Combined filter
mask_target   = mask_female & mask_1st_cls
df_target     = df[mask_target].copy()

print(f"Total passengers              : {len(df)}")
print(f"Female AND 1st class          : {len(df_target)}")
print(f"As % of total dataset         : {len(df_target)/len(df)*100:.1f}%")
print()
df_target[['survived', 'pclass', 'sex', 'age', 'fare', 'embarked']].head(10)
````

````python
# Cross-tab to verify the filter is correct
print("Cross-tab: sex × pclass (counts)")
print(pd.crosstab(df['sex'], df['pclass'], margins=True))
````

---
## Task 2 — Survival Rate for This Group

`survived` is already 0/1, so `mean()` directly gives the survival rate.

````python
# --- Survival rate for female 1st-class passengers ---
target_survival_rate = df_target['survived'].mean()
target_survived      = df_target['survived'].sum()
target_died          = len(df_target) - target_survived

print("GROUP: Female passengers in 1st class")
print(f"  Total in group     : {len(df_target)}")
print(f"  Survived           : {int(target_survived)}")
print(f"  Did not survive    : {int(target_died)}")
print(f"  Survival rate      : {target_survival_rate:.4f}  ({target_survival_rate*100:.1f}%)")
````

---
## Task 3 — Compare with Overall Survival Rate

````python
# Overall survival rate
overall_rate = df['survived'].mean()

# Survival rates across all sex × class combinations
group_rates  = df.groupby(['sex', 'pclass'])['survived'].agg(['mean', 'count', 'sum'])
group_rates.columns = ['Survival Rate', 'Total', 'Survived']
group_rates['Survival Rate'] = group_rates['Survival Rate'].round(4)

print(f"Overall survival rate (all passengers): {overall_rate:.4f}  ({overall_rate*100:.1f}%)")
print()
print("Survival rate by sex × pclass:")
print(group_rates.to_string())
````

````python
# --- Side-by-side comparison ---
diff        = target_survival_rate - overall_rate
lift        = target_survival_rate / overall_rate

comparison = pd.DataFrame({
    'Group'         : ['Overall (891)', 'Female, 1st class (94)'],
    'Survival Rate' : [f"{overall_rate*100:.1f}%", f"{target_survival_rate*100:.1f}%"],
    'Survived'      : [int(df['survived'].sum()), int(target_survived)],
    'Total'         : [len(df), len(df_target)]
})
display(comparison)

print(f"\nAbsolute difference : +{diff*100:.1f} percentage points above overall")
print(f"Lift (ratio)        :  {lift:.2f}x  — this group survives {lift:.2f}x more often than average")
````

````python
import matplotlib.pyplot as plt

groups = [
    ('Overall\n(all 891)', overall_rate),
    ('Male\n3rd class',    df[(df['sex']=='male')   & (df['pclass']==3)]['survived'].mean()),
    ('Male\n2nd class',    df[(df['sex']=='male')   & (df['pclass']==2)]['survived'].mean()),
    ('Male\n1st class',    df[(df['sex']=='male')   & (df['pclass']==1)]['survived'].mean()),
    ('Female\n3rd class',  df[(df['sex']=='female') & (df['pclass']==3)]['survived'].mean()),
    ('Female\n2nd class',  df[(df['sex']=='female') & (df['pclass']==2)]['survived'].mean()),
    ('Female\n1st class',  target_survival_rate),
]

labels = [g[0] for g in groups]
rates  = [g[1] for g in groups]
colors = ['steelblue' if 'Overall' in l
          else 'salmon' if 'Male' in l
          else 'mediumseagreen'
          for l in labels]

fig, ax = plt.subplots(figsize=(10, 5))
bars = ax.bar(labels, rates, color=colors, edgecolor='white', linewidth=0.8)

# Annotate bars with % value
for bar, rate in zip(bars, rates):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.015,
            f'{rate*100:.1f}%', ha='center', va='bottom', fontsize=10, fontweight='bold')

# Overall reference line
ax.axhline(overall_rate, color='navy', linestyle='--', linewidth=1.2, label=f'Overall avg ({overall_rate*100:.1f}%)')

ax.set_ylim(0, 1.1)
ax.set_ylabel('Survival Rate')
ax.set_title('Survival Rate by Sex × Passenger Class', fontsize=13, fontweight='bold')
ax.legend()
plt.tight_layout()
plt.show()
````

---
## Task 4 — ML Feature Engineering Insights

````python
print("""
INSIGHT 1: 'sex' and 'pclass' are among the strongest predictors of survival
─────────────────────────────────────────────────────────────────────────────
  Female 1st class survival rate : ~97%   ← nearly guaranteed survival
  Male   3rd class survival rate : ~14%   ← nearly guaranteed death
  Overall rate                   : ~38%

  → These two features alone separate most outcomes. Any model that drops
    'sex' or 'pclass' will have significantly lower accuracy.


INSIGHT 2: Interaction features capture what individual features miss
─────────────────────────────────────────────────────────────────────
  'sex'   alone: female = higher survival, but NOT uniformly.
  'pclass' alone: 1st class = higher survival, but NOT uniformly.

  The INTERACTION  sex × pclass  is far more informative:
    female_1st_class = 1   →  survival ~97%
    female_3rd_class = 1   →  survival ~50%  (large drop)

  FEATURE ENGINEERING ACTION:
    df['sex_class'] = df['sex'] + '_' + df['pclass'].astype(str)
    # Creates: 'female_1', 'female_2', 'female_3', 'male_1', 'male_2', 'male_3'
    # Then one-hot encode this combined feature.


INSIGHT 3: High survival in this subgroup = class imbalance within groups
─────────────────────────────────────────────────────────────────────────
  If you train a model ONLY on female 1st-class passengers, 97% of labels
  are '1' (survived). The model would predict '1' for everyone and still
  score 97% accuracy — but it learns nothing useful.

  → Always check survival rates per subgroup to detect local class imbalance.
  → Use stratified splits to ensure each class appears in train/test.


INSIGHT 4: Business logic → feature flags
─────────────────────────────────────────
  Create binary indicator columns for high-signal groups:

    df['is_female_1st'] = ((df['sex']=='female') & (df['pclass']==1)).astype(int)
    df['is_male_3rd']   = ((df['sex']=='male')   & (df['pclass']==3)).astype(int)

  These explicit flags make the interaction immediately available to
  linear models (which can't discover interactions on their own).
  Tree-based models discover interactions themselves, but explicit flags
  speed up convergence and improve interpretability.


INSIGHT 5: Filtering reveals the 'Women and children first' protocol
────────────────────────────────────────────────────────────────────
  Domain knowledge ≡ business logic: the Titanic evacuation followed a
  priority protocol. Encoding this as features (is_female, is_child,
  is_1st_class) directly captures the real-world decision process.
  This is what separates an ML practitioner from someone who just
  throws all columns at a model.
""")
````

````python
# --- Demonstrate the engineered features ---
df_engineered = df.copy()

# Interaction feature
df_engineered['sex_class']     = df['sex'] + '_' + df['pclass'].astype(str)

# Binary flags for high-signal groups
df_engineered['is_female_1st'] = ((df['sex'] == 'female') & (df['pclass'] == 1)).astype(int)
df_engineered['is_male_3rd']   = ((df['sex'] == 'male')   & (df['pclass'] == 3)).astype(int)
df_engineered['is_child']      = (df['age'] < 16).astype(int)

# Survival rate per sex_class group — confirms the engineered feature is powerful
print("Survival rate by engineered 'sex_class' interaction feature:")
print()
sex_class_stats = (
    df_engineered.groupby('sex_class')['survived']
    .agg(Survival_Rate='mean', Count='count')
    .sort_values('Survival_Rate', ascending=False)
    .assign(Survival_Rate=lambda x: x['Survival_Rate'].round(4))
)
print(sex_class_stats.to_string())

print()
print("New engineered columns added:")
print(df_engineered[['sex', 'pclass', 'age', 'sex_class', 'is_female_1st', 'is_male_3rd', 'is_child', 'survived']].head(10).to_string())
````
