# Handling Missing Data — Titanic

**Models cannot process null values.** Every `NaN` passed to `fit()` raises an error or silently corrupts predictions.  
Choosing the *right* strategy per column is as important as choosing the right model.

**Tasks:**
1. Fill missing `age` with **median**
2. Fill missing `embarked` with **mode**
3. Drop `deck` column entirely
4. Explain why different strategies suit different columns

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('titanic')

print(f"Shape: {df.shape}")
print()

# Missing values audit — always start here
missing = pd.DataFrame({
    'Missing Count' : df.isnull().sum(),
    'Missing %'     : (df.isnull().sum() / len(df) * 100).round(2),
    'Dtype'         : df.dtypes
}).query('`Missing Count` > 0').sort_values('Missing %', ascending=False)

print("Columns with missing values:")
print(missing.to_string())
```

```python
# Snapshot before changes
df_clean = df.copy()   
def missing_summary(dataframe, label=''):
    total = dataframe.isnull().sum().sum()
    cols  = (dataframe.isnull().sum() > 0).sum()
    print(f"[{label}]  Total NaNs: {total}  |  Columns with NaN: {cols}")

missing_summary(df_clean, 'Before cleaning')
```

```python
age_before = df_clean['age'].isnull().sum()
age_mean   = df_clean['age'].mean()
age_median = df_clean['age'].median()

print(f"Missing 'age' values   : {age_before}")
print(f"Mean age               : {age_mean:.2f} ")
print(f"Median age             : {age_median:.2f}")
print(f"Difference             : {abs(age_mean - age_median):.2f}")

# --- Fill with median ---
df_clean['age'] = df_clean['age'].fillna(age_median)

print()
print(f"Missing 'age' after fill : {df_clean['age'].isnull().sum()}")
```

```python
embarked_before = df_clean['embarked'].isnull().sum()
embarked_mode   = df_clean['embarked'].mode()[0]   # mode() returns a Series — take index 0

print(f"Missing 'embarked' values  : {embarked_before}")
print()
print("Value counts before fill:")
print(df_clean['embarked'].value_counts(dropna=False))
print()
print(f"Mode (most frequent port)  : '{embarked_mode}'")

# --- Fill with mode ---
df_clean['embarked'] = df_clean['embarked'].fillna(embarked_mode)

print()
print("Value counts after fill:")
print(df_clean['embarked'].value_counts(dropna=False))
print()
print(f"Missing 'embarked' after fill : {df_clean['embarked'].isnull().sum()}")
```

```python
# --- Drop the column ---
import seaborn as sns

df = sns.load_dataset('titanic')

df_clean = df.copy()
df_clean.drop(columns=['deck'], inplace=True)

print()
print(f"Columns before drop: {df.shape[1]}")
print(f"Columns after  drop: {df_clean.shape[1]}")
```
