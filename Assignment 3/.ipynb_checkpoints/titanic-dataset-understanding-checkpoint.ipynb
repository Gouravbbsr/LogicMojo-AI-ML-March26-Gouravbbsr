# Dataset Understanding — Titanic

The **first step** in any ML pipeline is dataset inspection.  
You cannot build a good model without understanding what's in your data.

**Tasks covered:**
1. `head()`, `tail()`, `info()`, `describe()`
2. Missing values per column
3. Numerical vs categorical features
4. Why feature types matter before modeling

```python
import pandas as pd
import numpy as np
import seaborn as sns

# Load Titanic dataset from seaborn
df = sns.load_dataset('titanic')

print("Dataset loaded successfully!")
print(f"Shape: {df.shape}  →  {df.shape[0]} rows, {df.shape[1]} columns")
print(f"Columns: {list(df.columns)}")
```

## Task 1D — `describe()` : Summary Statistics

`describe()` gives count, mean, std, min, quartiles, and max for **numerical** columns.  
Pass `include='all'` to also summarise **categorical** columns (top value, frequency, unique count).

```python
print("=== Numerical features only ===")
display(df.describe())

print("\n=== All features (numerical + categorical) ===")
display(df.describe(include='all'))
```

---
## Task 2A — Missing Values per Column

```python
total_rows = len(df)

missing = pd.DataFrame({
    'Missing Count'  : df.isnull().sum(),
    'Missing %'      : (df.isnull().sum() / total_rows * 100).round(2),
    'Dtype'          : df.dtypes
})

# Sort by missing count descending, show only columns with any missing
missing_only = missing[missing['Missing Count'] > 0].sort_values('Missing Count', ascending=False)

print(f"Total rows: {total_rows}")
print(f"Columns with missing values: {len(missing_only)} of {df.shape[1]}")
print()
print(missing_only.to_string())

print()
print("Interpretation:")
print("  'deck'     — 77% missing → too sparse to use; likely drop or engineer a 'has_deck' flag")
print("  'age'      — 20% missing → impute (median or model-based)")
print("  'embarked' —  0.2% missing → safe to fill with mode")
print("  'embark_town'/'alive' — derived from above; handle together")
```

---
## Task 2B — Numerical vs Categorical Features

```python
# Separate by dtype
numerical_cols   = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_cols = df.select_dtypes(include=['object', 'category', 'bool']).columns.tolist()

print(f"NUMERICAL features ({len(numerical_cols)}):")
for col in numerical_cols:
    print(f"  {col:<15} | dtype: {str(df[col].dtype):<10} | unique values: {df[col].nunique()}")

print()
print(f"CATEGORICAL features ({len(categorical_cols)}):")
for col in categorical_cols:
    unique_vals = df[col].dropna().unique()
    preview = str(list(unique_vals[:5]))
    print(f"  {col:<15} | dtype: {str(df[col].dtype):<10} | unique: {df[col].nunique()} | values: {preview}")
```

```python
# Visual summary table
summary_rows = []
for col in df.columns:
    dtype      = df[col].dtype
    n_unique   = df[col].nunique()
    n_missing  = df[col].isnull().sum()
    
    if dtype in ['int64', 'float64']:
        feat_type = 'Numerical'
    elif dtype == 'bool':
        feat_type = 'Boolean'
    elif dtype.name == 'category':
        feat_type = 'Categorical (ordered)'
    else:
        feat_type = 'Categorical'
    
    summary_rows.append({
        'Column'        : col,
        'Feature Type'  : feat_type,
        'Dtype'         : str(dtype),
        'Unique Values' : n_unique,
        'Missing'       : n_missing
    })

summary_df = pd.DataFrame(summary_rows).set_index('Column')
display(summary_df)
```

## Task 3 — Why Feature Types Matter Before Modeling

```python
print("""
WHY FEATURE TYPES MUST BE IDENTIFIED BEFORE MODELING
=====================================================

1. ENCODING REQUIREMENT
   ─────────────────────
   ML algorithms work with numbers — they cannot process raw strings.
   • Nominal categories (sex, embarked)  → One-Hot Encoding (creates 0/1 columns)
   • Ordinal categories (pclass, deck)   → Label / Ordinal Encoding (preserves order)
   • Boolean (survived, alone)           → Already 0/1, use as-is
   Applying the wrong encoder corrupts the data (e.g., label-encoding 'sex'
   as 0/1 implies male > female, which is meaningless).

2. SCALING REQUIREMENT
   ─────────────────────
   • Numerical features (age, fare)      → Must be scaled for distance-based
     models (KNN, SVM, Logistic Regression, Neural Nets).
   • Categorical features                → Scaling after one-hot encoding is
     usually skipped (values are already 0/1).
   Forgetting to scale age (0–80) vs fare (0–512) lets fare dominate KNN distance.

3. IMPUTATION STRATEGY
   ─────────────────────
   • Numerical missing (age)             → Mean / Median / KNN imputation
   • Categorical missing (embarked)      → Mode / 'Unknown' category
   • High-missing column (deck, 77%)     → Drop column or binary flag
   Using mean imputation on a categorical column is meaningless.

4. MODEL COMPATIBILITY
   ─────────────────────
   Model                 Numerical     Categorical
   ─────────────────────────────────────────────
   Linear / Logistic     raw or scaled  must encode first
   Decision Tree         raw is fine    must encode first
   CatBoost / LightGBM   raw is fine    native support (no encoding needed)
   Neural Network        normalize      one-hot or embedding

5. FEATURE IMPORTANCE & INTERPRETABILITY
   ─────────────────────────────────────
   Knowing which features are numerical lets you plot correlations, histograms,
   and box-plots. Categorical features need count-plots and chi-square tests.
   Mixing them up leads to wrong visualisations and misleading insights.

BOTTOM LINE: Misidentifying a feature type is one of the most common sources
of silent bugs in ML pipelines — the model trains without errors but produces
garbage predictions.
""")
```

```python
# Quick demo: what a pipeline looks like after proper type identification
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer

# Define which columns go through which transformer
num_features = ['age', 'fare', 'sibsp', 'parch']
cat_features = ['sex', 'embarked', 'class']

num_pipeline = Pipeline([
    ('impute', SimpleImputer(strategy='median')),   # handle missing age/fare
    ('scale' , StandardScaler()),                   # normalise range
])

cat_pipeline = Pipeline([
    ('impute', SimpleImputer(strategy='most_frequent')),  # handle missing embarked
    ('encode', OneHotEncoder(handle_unknown='ignore')),   # string → 0/1 columns
])

preprocessor = ColumnTransformer([
    ('numerical'  , num_pipeline, num_features),
    ('categorical', cat_pipeline, cat_features),
])

# Fit-transform on the dataframe
X_demo = df[num_features + cat_features]
X_transformed = preprocessor.fit_transform(X_demo)

print(f"Input shape  (raw)        : {X_demo.shape}")
print(f"Output shape (transformed): {X_transformed.shape}")
print()
print("Each categorical column expanded into one-hot columns.")
print("Numerical columns scaled. Missing values imputed.")
print("This is only possible because we identified feature types first.")
```
