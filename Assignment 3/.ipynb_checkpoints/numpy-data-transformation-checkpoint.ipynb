# Q10. Data Transformation for Model Input (NumPy)
**Context:** ML models require structured numeric arrays.

```python
import numpy as np

# Task 1: Create two arrays of shape (3, 2)
A = np.array([[1, 2],
              [3, 4],
              [5, 6]])

B = np.array([[7,  8],
              [9, 10],
              [11, 12]])

print('Array A (3x2):')
print(A)
print('\nArray B (3x2):')
print(B)
```

## Task 2a: Vertical Stacking — `np.vstack`
Stacks arrays row-by-row. Both arrays must have the **same number of columns**.

```python
v_stacked = np.vstack([A, B])
print('Vertical Stack (vstack) shape:', v_stacked.shape)  # (6, 2)
print(v_stacked)
```

## Task 2b: Horizontal Stacking — `np.hstack`
Stacks arrays column-by-column. Both arrays must have the **same number of rows**.

```python
h_stacked = np.hstack([A, B])
print('Horizontal Stack (hstack) shape:', h_stacked.shape)  # (3, 4)
print(h_stacked)
```

## Task 3: Reshape vstack result into (2, 6)

```python
reshaped = v_stacked.reshape(2, 6)
print('Original vstack shape :', v_stacked.shape)  # (6, 2) → total elements = 12
print('Reshaped to (2, 6)     :', reshaped.shape)   # (2, 6) → total elements = 12
print()
print('Reshaped array:')
print(reshaped)
print()
# verify no data was lost
print('Total elements before:', v_stacked.size)
print('Total elements after :', reshaped.size)
```

## Visual Summary of Operations

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

fig, axes = plt.subplots(1, 4, figsize=(14, 3))

def plot_array(ax, arr, title, cmap='Blues'):
    im = ax.imshow(arr, cmap=cmap, aspect='auto')
    ax.set_title(f'{title}\nshape={arr.shape}')
    for i in range(arr.shape[0]):
        for j in range(arr.shape[1]):
            ax.text(j, i, str(arr[i, j]), ha='center', va='center', fontsize=11, fontweight='bold')
    ax.set_xticks([]); ax.set_yticks([])

plot_array(axes[0], A, 'Array A', cmap='Blues')
plot_array(axes[1], B, 'Array B', cmap='Oranges')
plot_array(axes[2], v_stacked, 'vstack(A, B)', cmap='Purples')
plot_array(axes[3], reshaped, 'reshape(2, 6)', cmap='Greens')

plt.suptitle('NumPy Array Transformations', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

## Task 4: Why Reshaping Is Required in ML Pipelines

| Scenario | Reshape Need |
|---|---|
| **Flattening images** | A 28×28 pixel image → reshape to `(784,)` before feeding into a Dense layer |
| **Batch input** | Single sample `(features,)` → `(1, features)` for models expecting 2D input |
| **CNN input** | `(samples, H, W)` → `(samples, H, W, 1)` to add channel dimension |
| **stacking multiple arrays** | After `vstack` / `hstack`, shape may not match model's `input_shape` — reshape fixes this |
| **sklearn requirement** | `sklearn` models expect `X` as 2D `(n_samples, n_features)` — a 1D array must be reshaped |

**Core principle:**  
> Reshape **never changes data values or order** — only the dimensional view. It is a zero-cost memory operation in NumPy (returns a view, not a copy when possible). This makes it the standard tool for adapting array shapes to model API expectations without data loss.
