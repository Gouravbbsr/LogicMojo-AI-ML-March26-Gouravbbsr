# Data Selection for Model Input — Iris Dataset

Models are trained on **selective features**, not full raw data.  
This notebook covers NumPy slicing, boolean masking, and their role in ML pipelines.

**Tasks:**
1. Extract first 100 samples + last 2 features
2. Boolean masking — select samples where petal length > dataset mean
3. Count selected samples
4. Relate this to feature-based filtering in ML

````python
import numpy as np
from sklearn.datasets import load_iris

iris = load_iris()
X = iris.data          # shape (150, 4)
y = iris.target        # shape (150,)
feature_names = iris.feature_names

print("Full dataset shape :", X.shape)
print("Feature names      :", feature_names)
print("Classes            :", iris.target_names)
````

---
## Task 1 — Extract: First 100 samples, Last 2 features

NumPy slicing syntax: `array[row_slice, col_slice]`

| Slice | Meaning |
|---|---|
| `[:100]` | rows 0–99 (first 100 samples) |
| `[:, -2:]` | last 2 columns (petal length, petal width) |

````python
# Select first 100 rows AND last 2 columns simultaneously
X_selected = X[:100, -2:]

print("Selected features  :", feature_names[-2:])   # petal length, petal width
print("Shape after slicing:", X_selected.shape)      # (100, 2)
print()
print("First 5 rows:")
print(X_selected[:5])
print()
print("Last 5 rows:")
print(X_selected[-5:])
````

---
## Task 2 — Boolean Masking: petal length > dataset mean

A **boolean mask** is an array of `True/False` values with the same length as the dataset.  
When applied to an array, NumPy returns only the rows where the mask is `True`.

```
condition  →  [False, True, True, False, ...]   ← boolean mask
X[mask]    →  rows where condition is True
```

````python
# Petal length is column index 2 in the full dataset
petal_length = X[:, 2]

# Compute dataset-wide mean of petal length
mean_petal_length = np.mean(petal_length)
print(f"Mean petal length : {mean_petal_length:.4f} cm")

# --- Build the boolean mask ---
mask = petal_length > mean_petal_length

print(f"\nMask shape        : {mask.shape}")
print(f"Mask dtype        : {mask.dtype}")
print(f"First 10 mask vals: {mask[:10]}")
print(f"  (False = below mean, True = above mean)")
````

````python
# --- Apply mask to full dataset ---
X_filtered = X[mask]
y_filtered = y[mask]

print("Shape before masking:", X.shape)
print("Shape after  masking:", X_filtered.shape)
print()
print("First 5 filtered rows (all 4 features):")
print(X_filtered[:5])
````

---
## Task 3 — Count Selected Samples

````python
# Method 1: shape after masking
count_via_shape = X_filtered.shape[0]

# Method 2: sum the boolean mask (True = 1, False = 0)
count_via_sum = np.sum(mask)

# Method 3: np.count_nonzero
count_via_nonzero = np.count_nonzero(mask)

print(f"Total samples          : {X.shape[0]}")
print(f"Samples above mean     : {count_via_shape}  (via .shape)")
print(f"Samples above mean     : {count_via_sum}  (via np.sum)")
print(f"Samples above mean     : {count_via_nonzero}  (via np.count_nonzero)")
print(f"Samples at/below mean  : {X.shape[0] - count_via_shape}")
print()

# Class distribution in filtered set
print("Class distribution in filtered samples:")
for cls_id, cls_name in enumerate(iris.target_names):
    count = np.sum(y_filtered == cls_id)
    print(f"  {cls_name:<12}: {count} samples")
````

---
## Task 4 — How This Relates to Feature-Based Filtering in ML

What we just did manually mirrors several real ML pipeline stages:

````python
explanation = """
CONCEPT                   | WHAT WE DID                    | ML EQUIVALENT
--------------------------|--------------------------------|----------------------------------
Feature selection         | X[:, -2:]                      | SelectKBest, PCA, RFE
                          | (kept only petal features)     | — drop low-info columns before fit
                          |                                |
Sample filtering          | X[petal_length > mean]         | Outlier removal, class balancing
                          | (kept only informative rows)   | — IQR filter, SMOTE, threshold cut
                          |                                |
Train subset selection    | X[:100]                        | train_test_split, cross-validation
                          | (first 100 samples)            | — model never sees held-out rows
                          |                                |
Threshold-based decisions | petal_length > mean            | Precision/Recall thresholds,
                          | (boolean condition)            | confidence score cutoffs in prod
"""
print(explanation)

print("KEY INSIGHT — Why boolean masking matters in ML:")
print("""
  1. GARBAGE IN, GARBAGE OUT
     Feeding all 150 raw rows with all 4 features into a model is rarely optimal.
     Filtering by domain knowledge (e.g., 'petal length > threshold') removes
     noise and focuses the model on the informative region of feature space.

  2. CLASS IMBALANCE HANDLING
     The filtered set has 0 Setosa samples (all setosa petal lengths are well
     below the global mean). This is exactly the pattern used in anomaly detection
     — mask out the majority class before resampling.

  3. FEATURE SUBSETTING = DIMENSIONALITY REDUCTION
     Selecting only the last 2 features reduces the input space from 4-D to 2-D.
     Fewer dimensions = less risk of the curse of dimensionality, faster training.

  4. NO DATA LEAKAGE
     Always compute the mean (or any threshold) on the TRAINING set only,
     then apply the same mask to the test set. Computing the threshold on
     the full dataset leaks test information into training.
""")
````

````python
# --- Practical demo: train/test split + threshold from train only ---
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Compute mean petal length ONLY on training data
train_petal_mean = np.mean(X_train[:, 2])

# Apply same threshold to both splits
train_mask = X_train[:, 2] > train_petal_mean
test_mask  = X_test[:, 2]  > train_petal_mean   # <-- same threshold, not recomputed

print(f"Training mean petal length : {train_petal_mean:.4f} cm")
print(f"Train samples above mean   : {np.sum(train_mask)} / {len(X_train)}")
print(f"Test  samples above mean   : {np.sum(test_mask)}  / {len(X_test)}")
print()
print("Threshold derived from training data only — no data leakage.")
````
