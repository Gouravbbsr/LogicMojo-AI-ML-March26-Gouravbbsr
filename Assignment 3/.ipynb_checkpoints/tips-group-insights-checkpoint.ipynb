# Q8. Group-Based Insights (Customer Segmentation Logic)
**Context:** ML models often depend on aggregated patterns.
**Dataset:** Tips

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')
print('Shape:', tips.shape)
tips.head()
```

## Task 1a: Average Tip by Day

```python
avg_tip_by_day = tips.groupby('day')['tip'].mean().round(2)
print('Average Tip by Day:')
print(avg_tip_by_day)

fig, ax = plt.subplots(figsize=(7, 4))
avg_tip_by_day.plot(kind='bar', color='steelblue', edgecolor='black', ax=ax)
ax.set_title('Average Tip by Day')
ax.set_xlabel('Day')
ax.set_ylabel('Average Tip ($)')
ax.set_xticklabels(avg_tip_by_day.index, rotation=0)
for i, v in enumerate(avg_tip_by_day):
    ax.text(i, v + 0.02, f'${v}', ha='center', fontweight='bold')
plt.tight_layout()
plt.show()
```

## Task 1b: Total Bill by Gender

```python
total_bill_by_gender = tips.groupby('sex')['total_bill'].agg(['sum', 'mean', 'count']).round(2)
total_bill_by_gender.columns = ['Total Bill ($)', 'Avg Bill ($)', 'Num Transactions']
print('Total Bill Summary by Gender:')
print(total_bill_by_gender)

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

total_bill_by_gender['Total Bill ($)'].plot(kind='bar', ax=axes[0], color=['salmon', 'skyblue'], edgecolor='black')
axes[0].set_title('Total Bill by Gender')
axes[0].set_ylabel('Total Bill ($)')
axes[0].set_xticklabels(total_bill_by_gender.index, rotation=0)

total_bill_by_gender['Avg Bill ($)'].plot(kind='bar', ax=axes[1], color=['salmon', 'skyblue'], edgecolor='black')
axes[1].set_title('Average Bill by Gender')
axes[1].set_ylabel('Avg Bill ($)')
axes[1].set_xticklabels(total_bill_by_gender.index, rotation=0)

plt.tight_layout()
plt.show()
```

## Task 2: Create New Feature — tip_percentage

```python
tips['tip_percentage'] = (tips['tip'] / tips['total_bill'] * 100).round(2)

print('tip_percentage stats:')
print(tips['tip_percentage'].describe().round(2))
print()
print('Sample rows with new feature:')
tips[['total_bill', 'tip', 'tip_percentage']].head(8)
```

## Task 3: Which Day Has the Highest Tipping Behavior?

```python
day_tipping = tips.groupby('day').agg(
    avg_tip_pct=('tip_percentage', 'mean'),
    avg_tip_amt=('tip', 'mean'),
    num_customers=('tip', 'count')
).round(2)

best_day = day_tipping['avg_tip_pct'].idxmax()
print(f'Day-wise Tipping Summary:')
print(day_tipping)
print(f'\nHighest Tipping Day: {best_day} ({day_tipping.loc[best_day, "avg_tip_pct"]}% avg tip)')

fig, ax = plt.subplots(figsize=(7, 4))
bars = ax.bar(day_tipping.index, day_tipping['avg_tip_pct'],
              color=['gold' if d == best_day else 'lightgray' for d in day_tipping.index],
              edgecolor='black')
ax.set_title('Average Tip Percentage by Day\n(Gold = Highest Tipping Day)')
ax.set_xlabel('Day')
ax.set_ylabel('Avg Tip %')
for bar, val in zip(bars, day_tipping['avg_tip_pct']):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.1,
            f'{val}%', ha='center', fontweight='bold')
plt.tight_layout()
plt.show()
```

## Task 4: How This Helps in Segmentation Models

**Customer Segmentation Using Group Patterns:**

| Insight | Segmentation Use |
|---|---|
| `avg_tip_by_day` | Identify **high-revenue days** → weight training data or build day-specific models |
| `total_bill_by_gender` | Detect **spending behavior differences** → separate clusters for Male/Female segments |
| `tip_percentage` | Normalizes tip relative to bill → **fair comparison** across table sizes |
| Best tipping day | **Temporal feature** for models → Sunday/Saturday patterns reveal leisure vs workday behavior |

**Key ML takeaway:**  
- Raw `tip` amount is biased by `total_bill` — `tip_percentage` is a better feature for clustering and classification.  
- Group aggregates (`mean tip per day`) serve as **population-level priors** in Bayesian or tree-based models.  
- Day and gender become categorical features that split the feature space, enabling **targeted segment models** (e.g., weekend high-spenders vs weekday budget customers).
